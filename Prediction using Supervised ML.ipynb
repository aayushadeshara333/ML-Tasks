### Task-1 : Prediction using Supervised ML

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('http://bit.ly/w-data')
df.head()
```

```python
plt.scatter(x=df['Hours'], y=df['Scores'])
plt.title('Relation Between Hours And Scores', fontdict={'weight' : 'bold', 'size' : 18})
plt.xlabel('Hours')
plt.ylabel('Scores')
plt.show()
```

```python
X = df[['Hours']].values
Y = df[['Scores']].values
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=0)
```

## Training Model

```python
model = LinearRegression()
model.fit(x_train, y_train)
```

```python
y_predict = model.predict(x_test)
y_predict
```

## Staight line equation : y = mx + c

```python
l = (model.coef_ * X) + model.intercept_
plt.scatter(X, Y)
plt.plot(X, l, color='g')
plt.title('Linear regression visualizer', fontdict={'weight' : 'bold', 'size' : 18})
plt.xlabel('Hours')
plt.ylabel('Scores')
plt.show()
```

## Prediction

```python
prediction = model.predict([[9.25]])
print('Input in Hours :', 9.25)
print('Predicted output of Score :',prediction[0][0])
```

## Finding mean squared error

```python
from sklearn.metrics import mean_squared_error
plt.scatter(x_test, y_test, label = 'True Values')
plt.scatter(x_test, y_predict, label = 'Predicted Values')
plt.xlabel('Hours')
plt.ylabel('Scores')
plt.legend()
plt.show()
print('Mean Squared Error : ', mean_squared_error(y_test, y_predict))
```
